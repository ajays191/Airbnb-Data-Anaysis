#### Chart - 1

````python
# Chart - 1 visualization code
# The distribution of prices for each room type
data = {
    'room_type': ['Entire home/apt'] * 100 + ['Private room'] * 150 + ['Shared room'] * 50,
    'price': np.random.randint(50, 500, size=300)
}

df = pd.DataFrame(data)

# Calculate quartiles and potential outliers for each room type
room_types = df['room_type'].unique()
outliers = {}

for room_type in room_types:
    subset = df[df['room_type'] == room_type]
    quartiles = np.percentile(subset['price'], [25, 50, 75])
    iqr = quartiles[2] - quartiles[0]
    lower_bound = quartiles[0] - 1.5 * iqr
    upper_bound = quartiles[2] + 1.5 * iqr

    potential_outliers = subset[(subset['price'] < lower_bound) | (subset['price'] > upper_bound)]
    outliers[room_type] = potential_outliers

    print(f"Room Type: {room_type}")
    print(f"    25th Percentile (Q1): {quartiles[0]}")
    print(f"    Median (Q2): {quartiles[1]}")
    print(f"    75th Percentile (Q3): {quartiles[2]}")
    print(f"    IQR: {iqr}")
    print(f"    Potential Outliers:")
    print(potential_outliers)
    print("=" * 40)

# Visualization using box plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type')
plt.show()

print("Quartiles, Medians, Potential Outliers calculated, and visualization displayed.")
````

##### 1. Why did you pick the specific chart?

A box plot visually represents the distribution of data, including median, quartiles, and potential outliers. It's effective for comparing the spread of prices across different room types.

 In this project, we delve into the pricing trends of different types of accommodations: "Entire home/apt," "Private room," and "Shared room." **Key Findings:**
